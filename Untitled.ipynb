```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Understanding the Dataset:

```python
# import dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
        # make pandas dataframe and store in variable df 
df = pd.read_csv(url)    
print(df)
```

```python
# view the dimensions of dataset
print(df.shape)
```

```python
#check if any null value in dataset
print(df.info())
```

as from above data info it is clear that there are 244 rows
and all seven columns has 244 non-null values 
so there is no any null value entered in our dataset

from above data it is clear that we have five independent variables( sex, smoker, day, time, size) these are our features
and we have two dependent variables( total_bill, tip) these are our labels

```python
# define independent variables
x1=df['sex'].values
x2=df['smoker'].values
x3=df['day'].values
x4=df['time'].values
x5=df['size'].values
```

```python
#define independent variables
y1=df['total_bill'].values
y2=df['tip'].values
```

```python
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(x5, y1)
#print(r)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x5))

plt.scatter(x5, y1, color='black')
plt.plot(x5, mymodel, color='blue')
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x5, y2)
#print(r)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x5))

plt.scatter(x5, y2, color='black')
plt.plot(x5, mymodel, color='blue')
plt.show()
```

from above graphs we come to know that there is possible linear relationship between x5(size) and y1(total_bill)
so we design our linear model using these two parameters. now we asign size as 'X' and total_bill as 'Y'

```python
X=df['size'].values
Y=df['total_bill'].values
```

```python
#print dimensions of X and Y
print(X.shape)
print(Y.shape)
```

```python
#reshape dimensions
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
print(X.shape)
print(Y.shape)
```

# Model Building:
        splitting dataset intotraining and test dataset
  ## Training set
            X_train
            Y_train
  ## Testing set
            X_test
            Y_test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print('X_train = ', X_train.shape)
print('Y_train = ' , Y_train.shape)
print('X_test = ' , X_test.shape)
print('Y_test = ' , Y_test.shape)
```

 from above dimensions it is clear that we use 195 data enteries to train our model and 
 then we use remaining 49 data enteries are used to test the model

 Now build the simple linear regression model using training data sets

```python
from sklearn.linear_model import LinearRegression
l_model=LinearRegression()

# train model using training data set X_train and Y_train
l_model.fit(X_train, Y_train)

# make prediction from model on X_test data and store in Y_predict
Y_predict=l_model.predict(X_test)
print(Y_predict)
```

# Model Evaluation:
    1. Root Mean Square Error(rmse)
    2. R-squared

```python
# calculate rmse
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
# calculate R-squared
from sklearn.metrics import r2_score
print ("R2 Score value: {:.4f}".format(r2_score(Y_test, Y_predict)))
```
